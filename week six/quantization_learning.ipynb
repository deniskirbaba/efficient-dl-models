```python
import numpy as np

rng = np.random.default_rng(1)

X = rng.choice(np.linspace(-0.5, 10.0, 1_000), size=(128, 256))
print(X.min(), X.max())

quant_range = (-128, 127)
print(quant_range[0], quant_range[1])
```

```python
xmin = X.min()
xmax = X.max()
qmin, qmax = quant_range

scale = (xmax - xmin) / (qmax - qmin)
zero_point = int(round(qmin - xmin / scale))

inp = 0.0
q = int(round(inp / scale + zero_point))
q = np.clip(q, qmin, qmax)

print(f"{scale=}, {zero_point=}, {q=}")
```
